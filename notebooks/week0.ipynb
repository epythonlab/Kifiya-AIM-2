We have removed any negative values in each datasets of the specified columns.

Next, let's check and detect outliers, especially in sensor readings (ModA, ModB) and wind speed data (WS, WSgust) if any.

There are different approaches to detect outliers such as box-plot, z-score, and interquantil range. But now, I am going to use Interquantil range(IQR) to check and avoid outliers.


```python
# Define a function to identify outliers using IQR
def find_outliers(df, columns):
    outliers_dict = {}
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers_dict[column] = outliers
    return outliers_dict
```

```python
# List of dataset names and dataframes
datasets = {
    'Togo': togo_df,
    'Sierra Leone': seralio_df,
    'Benin': benin_df
}

# Check for outliers
for name, df in datasets.items():
    print(f"Outliers in {name} dataset:")
    outliers_iqr = find_outliers(df, ['ModA', 'ModB', 'WS', 'WSgust'])
    for column, outliers in outliers_iqr.items():
        print(f"  {column}: {len(outliers)} outliers")
        # Uncomment below to print the actual outliers
        # print(outliers)
    print()  # Print a blank line for better readability
```

The calculations are performed and the outliers are detected in the above summary output. Therefore, we can drop the outliers by selecting the data above the Lower Fence and below the Upper Fence values.

```python
def remove_outliers(df, columns):
    df_cleaned = df.copy()  # Create a copy to avoid modifying the original DataFrame
    for column in columns:
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    return df_cleaned

# List of dataset names and dataframes
datasets = {
    'Togo': togo_df,
    'Sierra Leone': seralio_df,
    'Benin': benin_df
}

# List of dataset names and DataFrames
datasets = {
    'Togo': togo_df,
    'Sierra Leone': seralio_df,
    'Benin': benin_df
}

# Remove outliers and update the original DataFrames
for name, df in datasets.items():
    # Remove outliers
    cleaned_df = remove_outliers(df, ['ModA', 'ModB', 'WS', 'WSgust'])
    
    # Update the original DataFrame
    if name == 'Togo':
        togo_df = cleaned_df
    elif name == 'Sierra Leone':
        seralio_df = cleaned_df
    elif name == 'Benin':
        benin_df = cleaned_df
    
    # Print summary
    print(f"Updated {name} dataset:")
    print(f"Original rows: {len(df)}")
    print(f"Updated rows: {len(cleaned_df)}")
    print()  # Print a blank line for better readability
```
